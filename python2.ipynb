1. Напишите скрипт, который проверяет наличие дубликатов в наборе данных и выводит информацию о количестве их обнаружения.

```python
def check_duplicates(data):

    # Подсчет количества дубликатов
    num_duplicates = len(data[data.duplicated()])

    # Вывод информации о дубликатах
    if num_duplicates > 0:
        print(f"В наборе найдено {num_duplicates} дубликатов")
    else:
        print("Дубликаты не обнаружены.")
```

```python
# Количество дубликатов в первом наборе
check_duplicates(set1)
```

```python
# Количество дубликатов во втором наборе
check_duplicates(set2)
```

```python
# Удалим дубликаты из наборов данных
set1 = set1.drop_duplicates()
set2 = set2.drop_duplicates()
```

2. Разработайте скрипт, который анализирует данные на наличие пропущенных и нулевых значений (там, где они не уместны) и предоставляет отчет о количестве пропусков и нулевых неуместных ячеек в каждом столбце.

Предположим, что нулевые значения уместны только в столбцах с типом float и int (когда речь идет о числовых значениях). Хотя каждый случай уникален и зависит от ситуации. В нашем примере name, registration_date нулевым быть не может.

```python
def check_nan_0(data):
    
    # Проверка пропусков в данных
    missing_values = data.isnull().sum()
    
    # Проверка на нулевые значения нечисловых столбцов
    invalid_zeros = (data.select_dtypes(exclude='number') == 0).sum()
    
    # Создаем отчет о пропусках и нулевых значениях
    report = pd.DataFrame({'Пропущенные значения': missing_values, 'Нулевые значения': invalid_zeros})
    
    return report
```

```python
# Проверка первого набора на пропуски и нулевые значения
check_nan_0(set1)
```

```python
# Проверка второго набора на пропуски и нулевые значения
check_nan_0(set2)
```

3. Напишите скрипт, который проверяет соответствие данных определенным форматам (например, даты, числа) и выводит список записей, не соответствующих этим форматам.

```python
# Проверка на формат даты. Функция ходит по значениям столбца и присваивает формат даты. 
# При возникновении ошибки преобразования в дату строка сохраняется в invalid_rows.

def get_invalid_date_rows(df, column_name):
    
    # Используем apply для проверки каждой строки
    invalid_rows = df[column_name].astype(str).apply(lambda x: pd.to_datetime(x, errors='coerce') is pd.NaT)
    
    # Возвращаем только строки с невалидными значениями
    return df[invalid_rows]
```

```python
# Поиск записей несоотвествующих дате из первого набора в столбце registration_date. 
get_invalid_date_rows(set1, 'registration_date')
```

```python
# Проверка на формат даты столбца с возрастом из первого набора. 
get_invalid_date_rows(set1, 'age')
```

```python
# Функция, проверяющая числовой формат значений в заданном столбце.

def check_numeric_column(df, column_name):
    return df[pd.to_numeric(df[column_name], errors='coerce').isna()]
```

```python
# Проверим столбец age на числовой формат в первом наборе данных.
check_numeric_column(set1, 'age')
```

```python
# Функция, проверяющая длину значения в столбце.

def check_column_length(dataframe, column_name, max_length):
    valid_rows = dataframe[dataframe[column_name].astype(str).apply(len) > max_length]
    return valid_rows
```

4. Создайте скрипт, который идентифицирует выбросы в данных, используя статистические методы, и предоставляет отчет о найденных аномалиях. Опишите, почему вами был выбран именно этот конкретный метод/методы, какие вы видите в нем достоинства и недостатки.

```python
# Выявление выбросов на основе метода Межквартильного размаха
def detect_outliers(data, column_name):
    
    # Вычисляем межквартильный размах
    Q1 = data[column_name].quantile(0.25)
    Q3 = data[column_name].quantile(0.75)
    IQR = Q3 - Q1

    # Границы для выявления выбросов
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Найдем выбросы
    outliers = data[(data[column_name] < lower_bound) | (data[column_name] > upper_bound)]

    print("Межквартильный размах (IQR):", round(IQR,2))
    print("Границы для определения выбросов:")
    print("Нижняя граница:", round(lower_bound,2))
    print("Верхняя граница:", round(upper_bound,2))
    print("Выбросы:")
    return outliers 
```

```python
# Применим данную функцию к столбцу income первого набора данных
detect_outliers(set1, 'income')
```

```python
# Выявление выбросов на основе метода 3х сигм
def find_outliers_std(df, column_name, threshold=3):
    
    # Рассчитаем среднее и стандартное отклонение
    mean_value = df[column_name].mean()
    std_dev = df[column_name].std()

    # Границы для выявления выбросов
    lower_bound = mean_value - threshold * std_dev
    upper_bound = mean_value + threshold * std_dev

    # Найдем выбросы
    outliers = df[(df[column_name] < lower_bound) | (df[column_name] > upper_bound)]

    # Результаты
    print("Среднее значение:", round(mean_value, 2))
    print("Стандартное отклонение:", round(std_dev, 2))
    print("Границы для определения выбросов:")
    print("Нижняя граница:", round(lower_bound, 2))
    print("Верхняя граница:", round(upper_bound,2))
    print("Выбросы:")
    return outliers
```

```python
find_outliers_std(set2, 'income', threshold=3)
```

При использовании метода на основе Межквартильного размаха за выбросы принимаются значения, находящиеся за пределами 1.5 величины межквартильного размаха от первого и третьего квартиля.
Метод 3 сигм основан на среднем значении и стандартном отклонении данных. Согласно этому методу 99.7% данных находятся в пределах 3 стандартных отклонениий от среднего значения. Наблюдения, находящиеся за пределами этого диапазона, считаются выбросами.
Метод на основе Межквартильного размаха более устойчив к экстремальным выбросам, так как основан на квартилях (а значит и медиане). В свою очередь, метод 3 сигм менее устойчив, так как опирается на среднее значение. В тоже время, метод с использованием Межквартильного размаха не учитывает форму распределения данных.
Мне кажется, что для выявления наиболее экстремальных выбросов лучше подойдет метод 3 сигм, так как данный метод оставляет только 0,3% наблюдений.

5. Напишите скрипт для проверки целостности связей между двумя наборами данных, выявляя несоответствия и противоречия в связанных данных.

```python
# Объединение 2х наборов данных по столбцам id и name
merged_df = pd.merge(set1, set2, on=['id', 'name'], how='outer', indicator=True)
merged_df.head(5)
```

```python
# Расчет количества строк во всех таблицах для сравнения
def check_sets(dataset1, dataset2, columns):
    print('Количество строк в первом наборе:', len(dataset1))
    print('Количество строк во втором наборе:', len(dataset2))
    merged_df = pd.merge(dataset1, dataset2, on=columns, how='outer', indicator=True)
    print('Количество совпавдающих строк в объединенной таблице:', len(merged_df[merged_df['_merge'] == 'both']))
    if len(merged_df[merged_df['_merge'] != 'both']) > 0:
        print('Несовпадающие строки при объединении наборов данных:')
        non_match = merged_df[merged_df['_merge'] !='both']
        return non_match
    else:
        print('Количество несовпадающих при объединении строк:', len(merged_df[merged_df['_merge'] !='both']))
```

```python
check_sets(set1, set2, ['id', 'name'])
```

```python
# Совпадение по столбцу age в 2х наборах
print('Количество пересечений по столбцу age:', len(merged_df[merged_df['age_x'] == merged_df['age_y']]))

# Совпадение по registration_date в 2х наборах
print('Количество пересечений по столбцу registration_date:', len(merged_df[merged_df['registration_date_x'] == merged_df['registration_date_y']]))

# Совпадение по income в 2х наборах
print('Количество пересечений по столбцу income:', len(merged_df[merged_df['income_x'] == merged_df['income_y']]))
```
